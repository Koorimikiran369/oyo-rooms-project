- We can see that '₹' is causing a problem in the Original_price and Final_price to convert the data type from object to float
- Also '% off' will not let us convert the Offer from object to float.  
- 'km' in the Distance will not let us convert Distance into float
- 'NEW' in Rating will not lets us convert the data from object to float for better analyses.
- 'ratings' in Number of Rating will not let us convert the data from object to float.


```python
# Removing '₹'from Original_price and Final_price
df['Original_price']=df['Original_price'].str[1:]
df['Final_price']=df['Final_price'].str[1:]
```

```python
df.head(5)
```

```python
# Removing '% off' from offer
df['Offer']=df['Offer'].str[:2]
```

```python
df.head(5)
```

```python
#Removing 'km' from Distance
df['Distance']=df['Distance'].str[:3]
```

```python
df.head(5)
```

```python
#Replacing 'NEW' with 0 for better analyses
df['Rating'].replace('NEW','0.0',inplace=True)
```

```python
df.head(15)
```

```python
#Removing () and ratings from Number of Rating
df['Number of Rating']=df['Number of Rating'].str[1:]
```

```python
df.head(5)
```

```python
df['Number of Rating']=df['Number of Rating'].str[:-8]
```

```python
df.head(5)
```

```python
df.rename(columns={'Number of Rating':'Number_of_Rating'},inplace=True)
```

```python
df.dtypes
```

```python
df['Original_price']=df['Original_price'].astype(float)
df['Offer']=df['Offer'].astype(float)
df['Final_price']=df['Final_price'].astype(float)
df['Distance']=df['Distance'].astype(float)
df['Rating']=df['Rating'].astype(float)
df['Number_of_Rating']=df['Number_of_Rating'].astype(float)
```

```python
df.dtypes
```

```python
df.drop_duplicates(inplace=True)
```
