```python
#read Hotel Names 

Name=[]

Hotel_name= soup.find_all(class_='listingHotelDescription__hotelName d-textEllipsis')

for j in range(len(Hotel_name)):

    if Hotel_name[j]:
        n = Hotel_name[j].text
    else:
        n=np.nan
    Name.append(n)
#Name
```

```python
Final_price= []

Hotel_finalprice= soup.find_all(class_='listingPrice__finalPrice')

for j in range(len(Hotel_name)):

    if Hotel_finalprice[j]:
        n = Hotel_finalprice[j].text
        n.replace('₹','')
    else:
        n=np.nan
    Final_price.append(n)
```

```python
Original_price=[]

Hotel_originalprice= soup.find_all(class_='listingPrice__slashedPrice d-body-lg')

for j in range(len(Hotel_name)):
    if Hotel_originalprice[j]:
        n = Hotel_originalprice[j].text
        n.replace('₹','')
    else:
        n=np.nan
    Original_price.append(n)
    
#Original_price    
```

```python
Offer=[]

Hotel_offer= soup.find_all(class_='listingPrice__percentage')

for j in range(len(Hotel_name)):
    if Hotel_offer[j]:
        n = Hotel_offer[j].text
    else:
        n=np.nan   
    Offer.append(n)
    
#Offer
```

```python
Total_rating=[]

Hotel_Total_rating= soup.find_all('div',class_='hotelRating')

for k in Hotel_Total_rating:
        x = k.text
        if len(re.findall(r'[(]\d+\s.*[)]',x)) == 1:
            Total_rating.append(re.findall(r'[(]\d+\s.*[)]',x)[0])
        else:
            Total_rating.append(np.nan) 

#Total_rating   
```

```python
a=[]
Summary=[]

for i in Hotel_Total_rating:
        a.append(i.text)

for i in a:
        #print(i)
        if i != 'NEW':
            Summary.append(i.split('·')[1])
        else:
            #Nur.append(np.nan) 
            Summary.append(np.nan)
#a
#Summary
```
