```python
import re
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

Name=[]
Final_price=[]
Original_price=[]
Offer=[]
Nur=[]
Rating=[]
Dist=[]
a=[]
Summary=[]


for x in range(10):
    ##########################Data Extraction#######################
    data= requests.get('https://www.oyorooms.com/hotels-in-btm-2nd-stage-bangalore/?page={}'.format(x))
    data_soup=BeautifulSoup(data.text,"html.parser")
    #####################Putting in HTML class for different attributes###################
    Hotel_name=data_soup.find_all(class_='listingHotelDescription__hotelName d-textEllipsis')
    Hotel_finalprice=data_soup.find_all(class_='listingPrice__finalPrice')
    Hotel_originalprice=data_soup.find_all(class_='listingPrice__slashedPrice d-body-lg')
    Hotel_offer=data_soup.find_all(class_='listingPrice__percentage')    
    Hotel_dist=data_soup.find_all(class_='listingHotelDescription__distanceText')    
    Hotel_Nur=data_soup.find_all('div',class_='hotelRating')
    Hotel_rating=data_soup.find_all('span',class_='hotelRating__rating')
            #####################Converting all attributes into lists#################
    for j in range(len(Hotel_name)): 

            if Hotel_name[j]:
                n = Hotel_name[j].text
            else:
                n=np.nan  
            Name.append(n)


            if Hotel_finalprice[j]:
                n = Hotel_finalprice[j].text
                n.replace('₹','')
            else:
                n=np.nan  
            Final_price.append(n)


            if Hotel_originalprice[j]:
                n = Hotel_originalprice[j].text
                n.replace('₹','')
            else:
                n=np.nan
            Original_price.append(n)


            if Hotel_offer[j]:
                n = Hotel_offer[j].text
            else:
                n=np.nan   
            Offer.append(n)


            if Hotel_dist[j]:
                n = Hotel_dist[j].text
                n.replace('km','')
            else:
                n=np.nan
            Dist.append(n)


            if Hotel_rating[j]:
                 n = Hotel_rating[j].text
            else:
                n=np.nan
            Rating.append(n)


    for k in Hotel_Nur:
        x = k.text
        if len(re.findall(r'[(]\d+\s.*[)]',x)) == 1:
            Nur.append(re.findall(r'[(]\d+\s.*[)]',x)[0])
        else:
            Nur.append(np.nan) 


    for i in Hotel_Nur:
        a.append(i.text)


    

for i in a:
        #print(i)
        if i != 'NEW':
            Summary.append(i.split('·')[1])
        else:
            #Nur.append(np.nan) 
            Summary.append(np.nan)

            
            
df=pd.DataFrame()
df['Name']=Name
df['Original_price']=Original_price
df['Offer']=Offer
df['Final_price']=Final_price
df['Distance']=Dist
df['Rating']=Rating
df['Number of Rating']=Nur
df['Summary']=Summary
df
```

```python
df.to_csv('BTM_2nd_Stage.csv',index=False,header=True)
```
