## Rating

```python
df['Rating'].describe()
```

```python
df['Rating'].value_counts()
```

```python
df['Rating'].value_counts().plot.bar()
```

```python
f,ax=plt.subplots(figsize=(10,9))
ax1=plt.subplot(211)
f.suptitle('Rating percentages, counts and distribution')
explode=(0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,1)
df['Rating'].value_counts(ascending=False).head(10).plot('pie',autopct="%0.2f%%",explode=explode,ax=ax1)

ax2=plt.subplot(223)
df['Rating'].plot.hist()
ax2.plot

ax3=plt.subplot(224)
sns.distplot(df['Rating'],ax=ax3)
```

### Observations

- 12.41% of the total properties are new or have 0 rating.
- 11.39% of the total properties have a rating of 4.1
- Only 6.93% of the total properties have a rating of 3.6
